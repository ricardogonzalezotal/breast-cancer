## RandomForest


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix

%run "../util/funciones.ipynb"
```

```python
#Carga de datos
df=pd.read_csv('../data/processed/output.csv')
df.head(10)
```

Train & Test

```python
X = df.drop('diagnosis',axis=1)
y = df['diagnosis']

x_train, x_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size = 0.25,
                                                   random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

RandomForestClassifier

```python
tree_r_clas = RandomForestClassifier(max_depth=3,random_state=42)
tree_r_clas.fit(x_train_scaled,y_train)
```

```python
predicciones_tree_r = tree_r_clas.predict(x_test_scaled)
print(classification_report(y_test, predicciones_tree_r))
```

```python
tree_r_clas.score(x_test_scaled,y_test)
```

```python
tree_r_clas.feature_importances_
```

```python
#Resultados validacion cruzada
score = ['f1','precision','accuracy','recall']
print('Resultados validacion cruzada')
for sc in score:
    rendimiento_tree_ran = cross_val_score(tree_r_clas, x_train_scaled, y_train, cv=5, scoring=sc)
    rendimiento_medio_tree_ran = rendimiento_tree_ran.mean()
    print('rendimiento medio',sc,':',rendimiento_medio_tree_ran)
```

```python
c_matrix_tree_r = confusion_matrix(y_test, predicciones_tree_r)
sns.heatmap(c_matrix_tree_r, annot=True)
print(c_matrix_tree_r)
```

Radomforest GridSearchCV y cross_val

```python
param_grid = {
    'max_depth': [5, 7, 10],
    'min_samples_split': [5, 7, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid_search = GridSearchCV(tree_r_clas, param_grid, cv=5)
grid_search.fit(x_train_scaled, y_train)
mejores_parametros_ran = grid_search.best_params_
mejor_modelo_ran = grid_search.best_estimator_
print('Resultados validacon cruzada')
for sc in score:
    rendimiento_ran = cross_val_score(mejor_modelo_ran, x_train_scaled, y_train, cv=5, scoring=sc)
    rendimiento_medio_ran = rendimiento_ran.mean()

    print('rendimiento medio',sc,':',rendimiento_medio_ran)
```
